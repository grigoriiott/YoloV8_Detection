```python
annotations_paths = glob.glob('../DATA/wide_imgs/img*.txt')
```

```python
classes_in = set()
for path in annotations_paths:
    f = open(path, 'r')
    l = [line[0] for line in f]
    for el in l:
        classes_in.add(el)
    f.close()
```

# Разобьем файлы на тестовую и тренировочную выборки

```python
from sklearn.model_selection import train_test_split
import shutil
```

```python
img_paths = glob.glob(data_path+'img*.jpg')
```

```python
img_paths.sort()
```

```python
annotations_paths.sort()
```

```python
img_paths[:5]
```

```python
annotations_paths[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(img_paths, annotations_paths, train_size=0.8)
```

```python
for i in range(len(X_train)):
    shutil.copy(X_train[i], '../DATA/train/images'+X_train[i][17:])
    shutil.copy(y_train[i], '../DATA/train/labels'+y_train[i][17:])
```

```python
for i in range(len(X_test)):
    shutil.copy(X_test[i], '../DATA/valid/images'+X_test[i][17:])
    shutil.copy(y_test[i], '../DATA/valid/labels'+y_test[i][17:])
```

```python
#Файлы были размещены по дирректориям train и valid
```

# Подготовка модели

Так как предоставленый датасет уже размечен в подходящем для yolo формате, для решения задачи детектирования будем использовать одну из моделей yolov8. Yolo позволяет достичь высоких результатов в точности детектирования, а также показывает хорошие результаты производительности в real-time.
Также в документации yolo предоставлены уже предобученные модели с метриками, благодаря чему можно осуществить выбор наиболее подходящей, исходя из требований.

Так как в ТЗ нет ограничений по требуемой точности выберем самую простую модель (с наименьшим количеством параметров) для уменьшения временных затрат на обучение модели

```python
from ultralytics import YOLO
```

```python
model = YOLO("../base_model/yolov8n.pt")
```

```python
model.info()
```

```python
history = model.train(
    data = "../DATA/data.yaml",
    epochs = 200,
    imgsz = 640,
    device = 'mps',
    patience = 30,
    batch = 16,
    save_period = 20,
    verbose = True,
    optimizer = 'Adam',
)
```
